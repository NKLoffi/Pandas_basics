# Pandas for analyzing


```python
import pandas as pd
```

```python
df = pd.read_csv('Data Jobs.csv')
```

```python
# Data Cleaning

df['job_posted_date']  = pd.to_datetime(df['job_posted_date'])
```

```python
india_jobs = df[df['job_country'] == 'India']
```

```python
india_jobs
```

```python
india_jobs.to_csv('india_jobs.csv', index = False) # make it into another csv file just to show the jobs from India
```

```python
india_jobs = india_jobs[india_jobs['salary_year_avg'].notna()] # Cleans the india_jobs DataFrame by removing rows with missing salary data
```

# Salary Statistics by Job Title in India (Sorted by Median Salary)


```python
india_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by = 'median', ascending = False)
```
