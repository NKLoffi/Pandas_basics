# Analyzing Data Science Job Demand

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading data 
df = pd.read_csv('/Users/nikhilaji/Documents/Pandas_basics/Pandas/Data Jobs.csv')
```

```python
# Data cleaning

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
```

```python
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')
```

```python
df_us_pivot = df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_us_pivot.reset_index(inplace=True)
df_us_pivot['month_number'] = pd.to_datetime(df_us_pivot['job_posted_month'], format='%B').dt.month
```

```python
df_us_pivot.sort_values('month_number', inplace=True)
df_us_pivot.set_index('job_posted_month', inplace=True)
df_us_pivot.drop(columns='month_number', inplace=True)
```

```python
df_us_pivot
```

```python
df_us_pivot.plot(kind='line')
```

```python
top_3 = df_us['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_us_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Joba in The US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.show()
```
