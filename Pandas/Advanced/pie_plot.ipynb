```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/nikhilaji/Documents/Pandas_basics/Pandas/Data Jobs.csv')
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=180, autopct='%1f%%')
plt.title('Work From Home (WFH) Status')
plt.ylabel(" ")
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='pie')
plt.title('Job Titles in the Data Set')
plt.ylabel(" ")
plt.show()
```

```python
fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home': 'Work from Home',
    'job_no_degree_mention': 'JobDegree Req',
    'job_health_insurance': 'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.f%%', labels= ['False', 'True'])
    ax[i].set_title(title)
    print(i, column, title)

plt.show()
```
